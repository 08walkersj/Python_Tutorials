# Maximizing data-ink ratio with matplotlib

I will show some examples of how to maximize data-ink ratio with matplotlib. By default, matplotlib uses a lot of ink. Removing ink can take a lot of extra lines of code, but it is possible. It is a better to start by sketching your idea for figure design and then code, instead of letting the code determine the design. 

## Bar chart example

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(10)
y = np.random.random(10)*100 # 10 random numbers between 0 and 100

fig, ax = plt.subplots()
ax.bar(x, y) 
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'grey')
# remove top and right borders:
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'grey')

# remove all borders:
for k in ['right', 'top', 'bottom', 'left']:
    ax.spines[k].set_visible(False)
    
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'grey')

for k in ['right', 'top', 'bottom', 'left']:
    ax.spines[k].set_visible(False)

# remove ticks from x axis:
ax.xaxis.set_ticks_position('none')

# and labels:
ax.xaxis.set_ticklabels([])

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'grey')
for k in ['right', 'top', 'bottom', 'left']:
    ax.spines[k].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.xaxis.set_ticklabels([])

# cut the bars with white lines:
ax.hlines(ax.get_yticks(), *ax.get_xlim(), color = 'white', linewidth = 3)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar(x, y, color = 'lightgrey')
for k in ['right', 'top', 'bottom', 'left']:
    ax.spines[k].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.xaxis.set_ticklabels([])
ax.hlines(ax.get_yticks(), *ax.get_xlim(), color = 'white', linewidth = 3)

plt.show()
```

# Scatter plot

```python
x = np.arange(21) # numbers from 0 to 20
y = x + np.random.normal(size = 21, loc = 0, scale = 2) # add gaussian noise

fig, ax = plt.subplots()
ax.scatter(x, y) 
plt.show()
```

```python
fig, ax = plt.subplots()

# use points with less ink
ax.scatter(x, y, s = 15, c = 'black') 
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s = 15, c = 'black') 

# REMOVE EVERYTHING
for k in ['right', 'top', 'bottom', 'left']:
    ax.spines[k].set_visible(False)
ax.xaxis.set_ticks_position('none')
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticks_position('none')
ax.yaxis.set_ticklabels([])

# but add custom borders
ax.plot([-5, -5], [y.min(), y.max()], 'k-', linewidth = .7)
ax.plot([x.min(), x.max()], [y.min()-5, y.min()-5], 'k-', linewidth = .7)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s = 15, c = 'black') 
plt.axis('off')

ax.plot([-5, -5], [y.min(), y.max()], 'k-', linewidth = .7)
ax.plot([x.min(), x.max()], [y.min()-5, y.min()-5], 'k-', linewidth = .7)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, s = 15, c = 'black') 
plt.axis('off')

# make a dot-dash-plot
ax.scatter([-5]*x.size, y, marker = '_', c = 'black', linewidth = .7)
ax.scatter(x, [y.min()-5]*y.size, marker = '|', c = 'black', linewidth = .7)
plt.show()
```

# Colors

Choose a color scale that is suited to the data

```python
from pysymmetry.visualization.polarsubplot import Polarsubplot
import pyamps

# get AMPS object and remove southern hemisphere
a = pyamps.AMPS(300, -5, -5, 23, 100)
mlat, mlt = a.scalargrid
mlat, mlt = mlat[:10000, :], mlt[:10000, :]
a.scalargrid = (mlat, mlt)
a.calculate_matrices()

# calculate upward and horizontal current densities:
ju = a.get_upward_current()
jh = a.get_total_current_magnitude()
```

```python
fig, axes = plt.subplots(ncols = 2, figsize = (12, 6))
# (notice the ink reduction)
paxes = [Polarsubplot(ax, minlat = 60, linestyle = '--', linewidth = .7, color = 'lightgrey') for ax in axes]

# plot horizontal current density with two different color maps:
paxes[0].contourf(mlat, mlt, jh, levels = 30)
paxes[1].contourf(mlat, mlt, jh, cmap = plt.cm.jet, levels = 30)

plt.show()
```

```python
fig, axes = plt.subplots(ncols = 2, figsize = (12, 6))
paxes = [Polarsubplot(ax, minlat = 60, linestyle = '--', linewidth = .7, color = 'lightgrey') for ax in axes]

# plot field-aligned currents:
paxes[0].contourf(mlat, mlt, ju, levels = 20)
paxes[1].contourf(mlat, mlt, ju, cmap = plt.cm.bwr, levels = 20)
plt.show()
```

```python
fig, axes = plt.subplots(ncols = 2, figsize = (12, 6))
paxes = [Polarsubplot(ax, minlat = 60, linestyle = '--', linewidth = .7, color = 'lightgrey') for ax in axes]

# make sure that the color scale is symmetric:
levels = np.linspace(-.95, .95, 20)

paxes[0].contourf(mlat, mlt, ju, levels = levels)
paxes[1].contourf(mlat, mlt, ju, cmap = plt.cm.bwr, levels = levels)
plt.show()
```
