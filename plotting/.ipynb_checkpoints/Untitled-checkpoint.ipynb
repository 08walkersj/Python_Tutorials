```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from pysymmetry.visualization.polarsubplot import Polarsubplot
import functools # needed to create the on click wrapper
```

```python
x, y= np.random.uniform(-100, 100, size=1000), np.random.uniform(-100, 100, size=1000)
```

```python
# This definition will be called everytime you click, the last argument is the click "event"
def onclick(fig, axes, event):
    ix, iy= event.xdata, event.ydata # Collect the click co-ordinates
    axes.scatter(ix, iy, color='red', size=10000)
    return
```

```python
fig= plt.figure()
ax= fig.add_subplot(111)
ax.scatter(x, y)
```

```python
onclick_wrapper=functools.partial(onclick, fig, ) # Create the wrapper
cid = fig.canvas.mpl_connect('button_press_event', onclick_wrapper)
```
