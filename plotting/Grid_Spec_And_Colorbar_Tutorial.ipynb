```python
x= np.random.uniform(10, 100, 10)
y= np.random.uniform(0, 1, 10)
z= np.random.uniform(3, 50, 10)
```

```python
fig= plt.figure()
gs=fig.add_gridspec(2,4, height_ratios=[1, 5], width_ratios= [10,10, 1, 1])
ax= fig.add_subplot(gs[:, 0])
s= ax.scatter(x, y, c=z, cmap='jet')
colorbar=fig.colorbar(s, ax=ax, orientation='vertical')
colorbar.set_label(label='z component'+r'$Akm^{-1}$', rotation=270, labelpad=20)
```

```python
fig= plt.figure()
gs=fig.add_gridspec(2,3, height_ratios=[1, 5], width_ratios= [10,10, 1])
ax1= fig.add_subplot(gs[:, 0])
ax2= fig.add_subplot(gs[:, 1])
cax= fig.add_subplot(gs[1, 2])
s= ax1.scatter(x, y, c=z, cmap='jet')
colorbar=fig.colorbar(s, cax=cax, orientation='vertical')
colorbar.set_label(label='z component '+r'$Akm^{-1}$', rotation=90, labelpad=0)
```

```python
fig= plt.figure()
gs=fig.add_gridspec(2,3, height_ratios=[5, 1], width_ratios= [10,10, 1])
ax1= fig.add_subplot(gs[0, :])
cax= fig.add_subplot(gs[1, :])
ax1.scatter(x, y, c=z, cmap='jet')
mappable=mpl.cm.ScalarMappable(norm = mpl.colors.Normalize(vmin=min(z), vmax= max(z)), cmap='jet')
colorbar=fig.colorbar(mappable, cax=cax, orientation='horizontal')
colorbar.set_label(label='z component '+r'$Akm^{-1}$', rotation=0)
```

```python
fig= plt.figure()
gs=fig.add_gridspec(2,3, height_ratios=[5, 1], width_ratios= [10,10, 1])
ax1= fig.add_subplot(gs[0, :2])
cax1= fig.add_subplot(gs[1, :])
cax2= fig.add_subplot(gs[0, -1])
s=ax1.scatter(x, y, c=z, vmin=15, vmax= 40, cmap='jet')
mappable=mpl.cm.ScalarMappable(norm = mpl.colors.Normalize(vmin=15, vmax= 40), cmap='jet')
colorbar1=fig.colorbar(mappable, cax=cax1, extend='both', orientation='horizontal')
colorbar1.set_label(label='z component '+r'$Akm^{-1}$', rotation=0)
colorbar2=fig.colorbar(s, cax=cax2, extend='both', orientation='vertical')
colorbar2.set_label(label='z component '+r'$Akm^{-1}$', rotation=270, labelpad=20)
```
