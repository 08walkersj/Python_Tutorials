```python
#cannot load pickled figures using standard backend in notebook
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating example plot
fig= plt.figure(num='Example')
theta= np.linspace(0, 2*np.pi)
plt.plot(theta, np.sin(theta), linestyle='--', marker='^', label='Sin(theta)', lw=1)
plt.plot(theta, np.cos(theta), linestyle=':', marker= '*', label='Cos(theta)', lw=1)
plt.fill_between(theta, np.sin(theta), np.cos(theta), color='darkred', label='area between', zorder=-1, alpha=.7)
legend_x=.6
legend_y=.8
plt.legend(loc='center', bbox_to_anchor=(legend_x, legend_y))
```

```python
#Save pickled figure
save_path= './'
file_name='Example'
import pickle
pickle.dump(fig, open(save_path+file_name+'.fig.pickle', 'wb'))
```

```python
#Reload pickled figure
figx = pickle.load(open(save_path+file_name+'.fig.pickle', 'rb'))
figx.show()
```
